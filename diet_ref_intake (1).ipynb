```python
def diet_ref_intake(age,sex,df):
    #function inputs characteristics of individual, output df.series of reccomended nutrients#
    #each specific column selected returns a Series#
    if age <= 3:
        col = 'C 1-3'
    age_range_list = [(4,8),(9,13),(14,18),(19,30),(31,50),(50,100)]
    for age_range in age_range_list:
        if age >= age_range[0] and age <= age_range[1]:
            col = sex + ' ' + str(age_range[0]) + '-' + str(age_range[1])
    return pd.Series(df[col])  
    
```
